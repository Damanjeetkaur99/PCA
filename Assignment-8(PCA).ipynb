```python
import pandas as pd 
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale 
```

```python
data=pd.read_csv("wine (1).csv")
df=pd.DataFrame(data)
df
```

```python
df1 = df.iloc[:,1:]
df1.head()
```

```python
wine = df1.values
wine
```

### Normalizing the data 

```python
wine_norm = scale(wine) 
wine_norm
```

### Transforming Data

```python
pca=PCA()
pca_values=pca.fit_transform(wine_norm)
```

```python
pca_values
```

### Amount of variance explained by each PCA

```python
var=pca.explained_variance_ratio_
var
```

## Cumulative Variance

```python
var1=np.cumsum(np.round(var,decimals=4)*100)
var1
```

```python
pca.components_
```

### Variance plot for PCA components obtained 

```python
plt.plot(var1,color='red')
```

```python
 pca_values[:,0:1]
```

## Plot between PCA1 and PCA2 

```python
x = pca_values[:,0:1]
y = pca_values[:,1:2]
plt.scatter(x,y)
```

```python
finalDf = pd.concat([pd.DataFrame(pca_values[:,0:2],columns=['pc1','pc2']), df], axis = 1)
finalDf
```

```python
import seaborn as sns
sns.scatterplot(data=finalDf,x='pc1',y='pc2',hue='Type')
```

### Looking at the above plot 3 clusters could be formed

# Performing Heirarchial and K-means Clustering to identify same number of clusters obtained by PCA

## Hierarchical Clustering

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sn
```

```python
wine_deng=sch.dendrogram(sch.linkage(wine_norm,method='single'))
```

```python
cc=AgglomerativeClustering(n_clusters=3,affinity="euclidean",linkage='single')
y_cc = cc.fit_predict(wine_norm)
```

```python
Clusters=pd.DataFrame(y_cc,columns=['Clusters'])
Clusters.head(10)
```

```python
df1['h_clusterid']=Clusters
df1
```

```python
hc_sorted=df1.sort_values("h_clusterid")
hc_sorted.tail(10)
```

## K Means

```python
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1, 11): #1 to 10
    kmeans = KMeans(n_clusters=i,random_state=0) 
    kmeans.fit(wine_norm)
    wcss.append(kmeans.inertia_)# inertia=twss
    
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```
